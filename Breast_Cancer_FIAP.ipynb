```python
# Importa a biblioteca pandas para melhor visualiza√ß√£o dos dados (utilizada j√° no primeiro bloco de c√≥digo)
import pandas as pd

# Importa a biblioteca Matplotlib, essencial para criar gr√°ficos e visualiza√ß√µes
# Foi utilizada a partir da cria√ß√£o e apresenta√ß√£o dos dados originais sobre a distribui√ß√£o dos resultados da coluna sobre diagn√≥sticos.
import matplotlib.pyplot as plt

# Importa a biblioteca Seaborn, que √© constru√≠da sobre o Matplotlib.
# O Seaborn fornece uma interface de alto n√≠vel para desenhar gr√°ficos estat√≠sticos.
import seaborn as sns
```

```python
# Ajusta a op√ß√£o de exibi√ß√£o do Pandas com o valor 'None' para remover o limite de colunas exibidos por padr√£o
# Isso porque a base de dados cont√©m 32 colunas e 569 linhas
pd.set_option('display.max_columns', None)

# Indica a URL do data set hospedado no Github
base_de_dados = 'https://raw.githubusercontent.com/marceloklotz/breast-cancer-fiap/refs/heads/main/breast-cancer.csv'

# Leitura e armazenamento do data set na vari√°vel df (data-frame)
df = pd.read_csv(base_de_dados)

# Abre a visualiza√ß√£o das primeiras 50 linhas da base de dados
df.head(50)
```

üîç **Visualiza√ß√£o amig√°vel das features da base de dados**

- A partir da visualiza√ß√£o da data set, notou-se que a feature "diagnosis" deve ser tratada como sendo a vari√°vel-alvo (target) para o modelo preditivo. Trata-se de uma classifica√ß√£o bin√°ria, onde os resultados poss√≠veis s√£o: benigno (B) ou maligno (M).

- Em seguida, realizamos a an√°lise da distribui√ß√£o da vari√°vel-alvo (diagnosis) na base de dados. A contagem revelou uma distribui√ß√£o de 357 tumores benignos (62,74%) e 212 tumores malignos (37,26%), conforme c√≥digo abaixo.

-  Essa distribui√ß√£o √© um desbalanceamento moderado, o que pode tornar a acur√°cia simples (ou accuracy) uma m√©trica enganosa, particularmente quando h√° um custo muito diferente para os erros.

**Nota explicativa:**
- A acur√°cia √© a propor√ß√£o de classifica√ß√µes corretas (verdadeiros positivos + verdadeiros negativos) em rela√ß√£o ao total de inst√¢ncias.

- Assim, neste cen√°rio, se o modelo proposto simplesmente classificar todos os tumores como benignos (a classe majorit√°ria, 62,74%), ele automaticamente teria uma acur√°cia de 62,74%. Esse valor parece bom, mas o modelo n√£o teria na verdade identificado nenhum caso maligno.

- Dessa forma cabe buscar m√©tricas que focam no desempenho da classe minorit√°ria ("maligno") e que penalizam o "falso negativo".

```python
# 1. Contar as ocorr√™ncias de cada valor na coluna 'diagnosis'
contagem_diagnosis = df['diagnosis'].value_counts()

# 2. Calcular a propor√ß√£o percentual
percentual_diagnosis = df['diagnosis'].value_counts(normalize=True) * 100

# 3. Criar um DataFrame para visualiza√ß√£o
# Concatena a contagem absoluta e o percentual em um √∫nico DataFrame
tabela_resumo = pd.DataFrame({
    'Contagem Absoluta': contagem_diagnosis,
    'Percentual (%)': percentual_diagnosis.round(2)
})

# 4. Apresentar os resultados
print("--- Resumo da Feature 'diagnosis' (Benigno/Maligno) ---")
print(tabela_resumo)
print("\nNota: 'B' = Benigno, 'M' = Maligno")
```

```python
# Configura√ß√µes para melhor visualiza√ß√£o.
sns.set_style("whitegrid")

# Define o tamanho da figura.
plt.figure(figsize=(7, 5))

# Cria o gr√°fico de barras usando 'diagnosis'.
# Otimiza√ß√£o: Passar 'diagnosis' para 'hue' e usar 'legend=False' elimina o FutureWarning.
ax = sns.countplot(
    x='diagnosis',
    data=df,
    order=['B', 'M'],
    hue='diagnosis',           # Vari√°vel para determinar a cor
    legend=False,              # Remove a legenda redundante
    palette={'B': 'teal', 'M': 'salmon'} # Paleta de cores personalizada para clareza
)

# --- Adicionando R√≥tulos de Porcentagem nas Barras ---

# 1. Obt√©m o n√∫mero total de amostras para o c√°lculo de porcentagem
total = len(df)

# 2. Itera sobre as barras para adicionar o texto (porcentagem)
for p in ax.patches:
    # Obt√©m a altura da barra
    height = p.get_height()

    # Calcula a porcentagem
    percentage = '{:.1f}%'.format(100 * height/total)

    # Adiciona o texto (porcentagem) acima de cada barra
    ax.text(p.get_x() + p.get_width()/2.,
            height + 5,
            percentage,
            ha="center",
            fontsize=12,
            fontweight='bold')

# Configura√ß√£o de t√≠tulos e legendas
plt.title('Distribui√ß√£o de Diagn√≥sticos (Benigno vs. Maligno)', fontsize=16)
plt.xlabel('Diagn√≥stico (B=Benigno, M=Maligno)', fontsize=12)
plt.ylabel('Contagem de Amostras', fontsize=12)

# Exibe o gr√°fico
plt.show()
```

üîç **An√°lise b√°sica sobre as features existentes na base de dados**

As features encontradas na base de dados s√£o tipicamente obtidas a partir de imagens digitalizadas de uma massa mam√°ria, o c√≥digo abaixo foi utilizado para traduzir o conjunto de vari√°veis.

Aqui foi poss√≠vel verificar que as features est√£o agrupadas em tr√™s categorias principais: mean (m√©dia), se (erro padr√£o) e worst (pior/maior).

-   **_mean**: foi utilizado para representar a m√©dia dos valores calculados para a
caracter√≠stica em quest√£o.

-  **_se**: foi utilizado para representar o erro padr√£o (ou desvio padr√£o do erro), que √© uma medida da varia√ß√£o nos valores da caracter√≠stica.

-   **_worst**: foi utilizado para representar o valor m√°ximo, pior ou mais extremo da caracter√≠stica.

Os modelos de Machine Learning mais indicados para criar um modelo preditivo de diagn√≥stico de c√¢ncer de mama (benigno ou maligno) com base nas features fornecidas s√£o aqueles que se destacam em problemas de classifica√ß√£o bin√°ria.
